```python
# Question 5
# In What month is the most crimes committed yearly?
```

```python
crimes_of_year_and_month = crimes.groupby(["MONTH", "YEAR"]).count()["OFFENSE_CODE"].unstack()
```

```python
avg_crimes = crimes.groupby(["YEAR", "MONTH"]).count()["OFFENSE_CODE"].mean()
print(avg_crimes)
```

```python
def style_negative(v, props = None):
    if v < avg_crimes:
        return props
    
s2 = crimes_of_year_and_month.style.applymap(style_negative, props='color:blue;')\
              .applymap(lambda v: 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None)
s2
```

```python
def style_negative(v, props = None):
    if v < avg_crimes:
        return props


avg_table = crimes_of_year_and_month.style.applymap(style_negative,  props = "color : lightgreen")
avg_table
```

```python
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
avg_table.apply(highlight_max, props = "color : black ; background-color : orange", axis = 0)
avg_table
```
